All arrivals follow a Poisson Process an travel to the network only through Note 1 with the arrival rate of 10. On leaving node 1, customers travel to node 2 with probability of $\frac{1}{2}$ ($p_{12}$) or to note 3 with probability $\frac{1}{2}$ ($p_{13}$). On leaving node 2, customers travel to node 3. On leaving node 3, all customers leave the network.

b. Yes, it is a feedforward acylic network

c. We have:

$\lambda_2 = \lambda_{02} + \lambda_1 p_{12}$

= $0 + 10*\frac{1}{2} = 5$

$\lambda_3 = \lambda_{03} + \lambda_1 p_{13} + \lambda_3 p_{23}$

= $0 + 10*\frac{1}{2} + 1*5= 10$

The conditions that the network has a steady-state distribution:

$\rho_2 = \frac{\lambda_2}{c_2 \mu_2} < 1$
=> $\rho_2 > \frac{\lambda_2}{\mu_2}$
<=> $\rho_2 > 5$ (1)

$\rho_3 = \frac{\lambda_3}{c_3 \mu_3} < 1$
=> $\rho_3 > \frac{\lambda_3}{\mu_3}$
<=> $\rho_3 > 10$ (2)

From (1) and (2), the minimum number of servers for nodes 2 and 3, namely c2 and c3 are 6 and 11, respectively that the network has a steady-state distribution.
